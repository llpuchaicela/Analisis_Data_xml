```python
import pandas as pd 
from xml.dom import minidom
from xml.etree.ElementTree import parse
import spacy as sp
nlp = sp.load('en_core_web_sm')
```

### Extraccion contenido country_CR

```python
# Leer country_CR
doc_CR = parse('Informacion_Categorizada/TASS2019_country_CR_train.xml')
doc_CR
```

```python
doc_CR = pd.read_xml('Informacion_Categorizada/TASS2019_country_CR_train.xml')
doc_CR
```

```python
lista_CR=[]
for item in doc_CR.iterfind('tweet'):
    lista_CR.append(item.findtext('content'))
print(lista_CR)
```

```python
df_CR = pd.DataFrame({'lista':lista_CR})
df_CR
```

```python
# Guardar contenido en csv
data_CR= df_CR.to_csv("Content_CR.csv")
```
