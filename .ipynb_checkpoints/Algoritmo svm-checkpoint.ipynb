```python
#  https://www.cienciadedatos.net/documentos/py24-svm-python.html#:~:text=M%C3%A1quinas%20de%20Vector%20Soporte%20(Vector,campo%20de%20la%20ciencia%20computacional.

# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np
import time

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from sklearn.metrics import classification_report


# Preprocesado y modelado
# ==============================================================================
from sklearn.svm import SVC

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
#DATOS
d = pd.read_csv("CSV_Informacion_Categorizada/cont_sent_CR_PE_MX_UY.csv", sep="|")
df= d.rename(columns={'Unnamed: 0':'index'})
df.head(3)

countW = []
# Conteo de palabras
for i in df.Content.tolist():
    countW.append(len(i.split()))
dfc = df.assign(countW = countW)
dfc.head(2)
```

```python
dfc['Sentiment'] = dfc.Sentiment.map({'NEU':3, 'P':1,'N':0,'NONE':4})
dfc.Content
```

```python
fig, ax = plt.subplots(figsize=(20,6))
ax.scatter(dfc.index, dfc.countW, c=dfc.Sentiment);
ax.set_title("Datos Total Sentimientos");
```

```python
# Creamos al vector caracterizado
stop =stopwords.words('spanish') 
count_vector = CountVectorizer(max_features=1000, lowercase=True, analyzer='word', stop_words= stop,ngram_range=(1,1))
train_vector= count_vector.fit_transform(dfc['Content'])
test_vector = count_vector.transform(dfc['Content'])
```

```python
# Creación del modelo SVM lineal
# ==============================================================================

modelo = SVC(C = 100, kernel = 'linear', random_state=123)
t1= time.time()
modelo.fit(train_vector, dfc['Sentiment'])
t2= time.time()
prediction= modelo.predict(test_vector)
t3= time.time()
time_modelo= t2-t1
time_prediction= t3-t2
```

```python
# Predicciones test
# ==============================================================================
predicciones = modelo.predict(test_vector)
predicciones
```

```python
prediction
modelo
```

```python
#Resultados de prediccion
print("Training time: %fs; Prediction time: %fs" % (time_modelo, time_prediction))
```

```python
# Matriz de confusión de las predicciones de test
# ==============================================================================
confusion_matrix = pd.crosstab(
    dfc['Sentiment'].ravel(),
    predicciones,
    rownames=['Real'],
    colnames=['Predicción']
)
confusion_matrix
```

```python
from sinfo import sinfo
sinfo()
```

```python
sentiment = classification_report(dfc['Sentiment'], prediction, output_dict=True)
print('positive: ', sentiment['1'])
print('negative: ', sentiment['0'])
print('neutro: ', sentiment['3'])
```
