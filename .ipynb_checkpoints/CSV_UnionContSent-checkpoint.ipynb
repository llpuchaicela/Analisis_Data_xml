```python
import pandas as pd 
from xml.dom import minidom
from xml.etree.ElementTree import parse
```

```python
# Leer country_CR
df1 = parse('Informacion_Categorizada/TASS2019_country_CR_train.xml')
df2 = parse('Informacion_Categorizada/TASS2019_country_PE_train.xml')
df3 = parse('Informacion_Categorizada/TASS2019_country_MX_train.xml')
df4 = parse('Informacion_Categorizada/TASS2019_country_UY_train.xml')
```

```python
df1 = pd.read_xml('Informacion_Categorizada/TASS2019_country_CR_train.xml')
df1.head(5)
```

```python
# Country CR
listaId= []
listaUser = []
listaCont=[]
listaDate = []
listaLang = []
listaSent=[]
for item in df1.iterfind('tweet'):
    listaId.append(item.findtext('tweetid'))
    listaUser.append(item.findtext('user'))
    listaCont.append(item.findtext('content'))
    listaDate.append(item.findtext('date'))
    listaLang.append(item.findtext('lang'))
    listaSent.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSent)
```

```python
# COUNTRY PE
listaId2= []
listaUser2 = []
listaCont2=[]
listaDate2 = []
listaLang2 = []
listaSent2=[]
for item in df2.iterfind('tweet'):
    listaId2.append(item.findtext('tweetid'))
    listaUser2.append(item.findtext('user'))
    listaCont2.append(item.findtext('content'))
    listaDate2.append(item.findtext('date'))
    listaLang2.append(item.findtext('lang'))
    listaSent2.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSent2)
```

```python
# COUNTRY MX
listaId3= []
listaUser3 = []
listaCont3=[]
listaDate3 = []
listaLang3 = []
listaSent3=[]
for item in df3.iterfind('tweet'):
    listaId3.append(item.findtext('tweetid'))
    listaUser3.append(item.findtext('user'))
    listaCont3.append(item.findtext('content'))
    listaDate3.append(item.findtext('date'))
    listaLang3.append(item.findtext('lang'))
    listaSent3.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSent3)
```

```python
# COUNTRY UY
listaId4= []
listaUser4 = []
listaCont4=[]
listaDate4 = []
listaLang4 = []
listaSent4=[]
for item in df4.iterfind('tweet'):
    listaId4.append(item.findtext('tweetid'))
    listaUser4.append(item.findtext('user'))
    listaCont4.append(item.findtext('content'))
    listaDate4.append(item.findtext('date'))
    listaLang4.append(item.findtext('lang'))
    listaSent4.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSent4)
```

```python
# Country CR
for i in range (len(listaSent)):
    print(listaId[i],listaUser[i],listaCont[i],listaDate[i],listaLang[i],listaSent[i])
```

```python
dfCR = pd.DataFrame({'Id':listaId,
                     'User':listaUser,
                     'Content':listaCont,
                     'Date':listaDate,
                     'Lang':listaLang,
                     'Sentiment':listaSent})
dfPE = pd.DataFrame({'Id':listaId2,
                    'User':listaUser2,
                    'Content':listaCont2,
                    'Date':listaDate2,
                    'Lang':listaLang2,
                    'Sentiment':listaSent2})
dfMX = pd.DataFrame({'Id':listaId3,
                    'User':listaUser3,
                    'Content':listaCont3,
                    'Date':listaDate3,
                    'Lang':listaLang3,
                    'Sentiment':listaSent3})
dfUY = pd.DataFrame({'Id':listaId4,
                    'User':listaUser4,
                    'Content':listaCont4,
                    'Date':listaDate4,
                    'Lang':listaLang4,
                    'Sentiment':listaSent4})
```

```python
# COUNTRY MX
listaContM=[]
listaSentM=[]
for item in df3.iterfind('tweet'):
    listaContM.append(item.findtext('content'))
    listaSentM.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSentM)
```

```python
# COUNTRY UY
listaContU=[]
listaSentU=[]
for item in df4.iterfind('tweet'):
    listaContU.append(item.findtext('content'))
    listaSentU.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSentU)
```

```python
dfC = pd.DataFrame({ 'Content':listaCont1,
                     'Sentiment':listaSent1})
dfP = pd.DataFrame({'Content':listaContP,
                    'Sentiment':listaSentP})
dfM = pd.DataFrame({'Content':listaContM,
                    'Sentiment':listaSentM})
dfU = pd.DataFrame({'Content':listaContU,
                    'Sentiment':listaSentU})
```

```python
dfcr= dfC.to_csv("CSV_Informacion_Categorizada/cont_sent_CR.csv", sep="|")
dfpe= dfP.to_csv("CSV_Informacion_Categorizada/cont_sent_PE.csv", sep="|")
dfmx= dfM.to_csv("CSV_Informacion_Categorizada/cont_sent_MX.csv", sep="|")
dfuy= dfU.to_csv("CSV_Informacion_Categorizada/cont_sent_UY.csv", sep="|")
```
